# Exercise 1

```python
import pandas as pd

# Load penguins dataset
df = pd.read_csv("penguins.csv")

# Display 5 rows
print("--------- First 5 rows: ---------")
print(df.head(5))

# Display info
print("\n--------- Dataframe info: ---------")
print(df.info())

# Count NaN values
print("\n--------- Count of NaN values --------- \n", df.isnull().sum().sum())
```

```python
# Select only rows with species == Adelie
df.loc[df['species'] == 'Adelie']
```

```python
# Create new Dataframe with only some columns
new_df = df[['species', 'bill_length_mm', 'body_mass_g']].copy()

# Filter penguins with weight > 4000g
print("--------- Filter penguins with weight > 4000g ---------")
print(new_df.loc[new_df['body_mass_g'] > 4000])

# Create new column
print("\n--------- Create new bill_ratio column ---------")
df["bill_ratio"] = ( df["bill_length_mm"] / df["bill_depth_mm"] )
print(df.columns)

# Convert sex column to uppercase
print("\n--------- Convert sex to uppercase ---------")
df['sex'] = df['sex'].str.upper() 
print(df['sex'].head(5))

# Replace null values in bill_length_mm column by column mean
print("\n--------- Replace null values by column mean ---------")
df["bill_length_mm"].fillna(df['bill_length_mm'].mean(), inplace = True)
print(df["bill_length_mm"])
```

```python
# Load UCI ML dataset

uci_data = pd.read_csv("iris_dataset/iris.data", names=["sepal_length_cm", "sepal_width_cm", "petal_length_cm", "petal_width_cm", "class"])
uci_data
```

# Exercise 3

```python
import seaborn as sns

# Line plot
sns.lineplot(uci_data, x="sepal_length_cm", y="sepal_width_cm")
```

```python
# Histogram
sns.histplot(uci_data, x="class", y="petal_length_cm")
```

```python
# Boxplot
sns.boxplot(uci_data, x="class", y="petal_width_cm")
```

```python
# Summary statistics
uci_data['sepal_length_cm'].describe()
```

```python
# Scatter plot
sns.scatterplot(uci_data, x="sepal_length_cm", y="petal_length_cm")
```

```python
# Scatter plot matrix with colors for each class
sns.pairplot(uci_data, hue="class")
```

```python
# Correlation matrix plot
sns.heatmap(uci_data.corr(numeric_only=True))
```
