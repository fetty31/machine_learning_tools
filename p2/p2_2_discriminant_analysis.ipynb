# PART 1: A FIRST SUPERVISED CLASSIFICATION PROBLEM WITH LDA AND QDA




## 1. Generate synthetic data (data matrix + class labels)

First, we are going to generate and visualize synthetic data for a binary classification problem. Execute the following lines of code to generate the data and visualize it.

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

X, y = make_classification(n_samples=360, n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, n_classes=2, random_state=42)

plt.figure(figsize=(7,7))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Data")
plt.show()
```

# 2. Split data in training and validation subsets

Separate the data in a training and validation subsets. Create one scatter plot for the training dataset and another for the validation dataset.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=46)

plt.figure(figsize=(7,7))
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train)
plt.title("Training Data")
plt.show()

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test)
plt.title("Testing Data")
plt.show()
```

# 3. Fit/train a LDA classification model

Train and validate a LDA model with your dataset. You may consider the following.

https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html

For the validation process, consider the following two steps:

1) Plot one scatter plot for the test data and another one for the predicted test data. Analyse the decision boundary of the algorithm. Why does the boundary take this form? Do you think that is a good decision boundary for the considered data?

2) Compute and analyse the confusion matrix.


```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

clf = LinearDiscriminantAnalysis()
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test)
plt.title("Test Data")
plt.show()

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred)
plt.title("Predicted Data")
plt.show()
```

```python
# Plot confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

C = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(C)
disp.plot()
```

Train and validate an QDA model with your dataset. You may consider the following.

# PART 2: A SECOND SUPERVISED CLASSIFICATION PROBLEM WITH LDA AND QDA

In this second part, we are going to generate and visualize synthetic data for a binary classification problem. Execute the following lines of code to generate the data.

```python
from sklearn.datasets import make_blobs,make_moons
import matplotlib.pyplot as plt

n_samples = 1500
centers = [(0, 0), (5, 5)]
X, y = make_moons(n_samples=n_samples,noise=0.4)

plt.figure(figsize=(7,7))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Data")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=46)

plt.figure(figsize=(7,7))
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train)
plt.title("Training Data")
plt.show()

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test)
plt.title("Testing Data")
plt.show()
```

Repeat the traning and validation process of a LDA and a QDA model in this new dataset. Compare the decision boundary of both algorithms and the performance of both algorithms. Is there any difference? Why?

```python
# Linear Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

clf = LinearDiscriminantAnalysis()
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test)
plt.title("Test Data")
plt.show()

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred)
plt.title("Predicted Data via LDA")
plt.show()

# Quadratic Discriminant Analysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

clf = QuadraticDiscriminantAnalysis()
clf.fit(x_train, y_train)

y_pred_q = clf.predict(x_test)

plt.figure(figsize=(7,7))
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred_q)
plt.title("Predicted Data via QDA")
plt.show()
```
