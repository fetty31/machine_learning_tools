First, we are going to generate and visualize synthetic data for a binary classification problem. Execute the following lines of code to generate the data and visualize it.

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

X, y = make_classification(n_samples=360, n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, n_classes=2, random_state=42)

plt.figure(figsize=(7,7))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Data")
plt.show()
```
